```python
# Import libraries
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv(r'C:\Users\EASYWAY TECH CENTER\Desktop\marketing campaign.csv')

# Display the first 5 rows
df.head()
```

```python
# Data Overview & Cleaning

# 1. Check the info of the dataset
df.info()

# 2. Check for missing values
print(df.isnull().sum())

# 3. Check the balance of our main target: conversion_status
print("\nConversion Status Counts:")
print(df['conversion_status'].value_counts())
print("\nConversion Rate (%):")
print(round(df['conversion_status'].value_counts(normalize=True) * 100, 2))
```

```python
# Question 1: What is the overall campaign success rate?
# Calculate the overall conversion rate
total_customers = df.shape[0]
converted_customers = df[df['conversion_status'] == 'converted'].shape[0]
conversion_rate = (converted_customers / total_customers) * 100

print(total_customers)
print(converted_customers)
print(conversion_rate)
```

```python
# Question 2: Which customer occupation has the highest conversion rate?
# Group data by occupation and calculate conversion rate for each
occupation_conversion = df.groupby('occupation')['conversion_status'].apply(
    lambda x: (x == 'converted').mean() * 100  # Calculate percentage
).round(2).sort_values(ascending=False)

print("Conversion Rate by Occupation (%):")
print(occupation_conversion)

# Find the top-performing occupation
top_occupation = occupation_conversion.idxmax()
top_rate = occupation_conversion.max()
print(f"\nThe occupation with the highest conversion rate is '{top_occupation}' at {top_rate}%.")
```

```python
# Question 3: What is the best month to run the campaign?
# Group data by month and calculate conversion rate
monthly_performance = df.groupby('call_month')['conversion_status'].apply(
    lambda x: (x == 'converted').mean() * 100
).round(2).sort_values(ascending=False)

print("Conversion Rate by Month (%):")
print(monthly_performance)

# Find the best month
best_month = monthly_performance.idxmax()
best_month_rate = monthly_performance.max()
print(f"\nThe best month to run the campaign is '{best_month}' with a {best_month_rate}% conversion rate.")
```

```python
# Question 4: How does call duration relate to conversion?
# Calculate average call duration for converted vs not converted customers
duration_analysis = df.groupby('conversion_status')['call_duration'].mean().round(2)

print("Average Call Duration (seconds):")
print(duration_analysis)

# Calculate the difference
conv_duration = duration_analysis['converted']
not_conv_duration = duration_analysis['not_converted']
difference = conv_duration - not_conv_duration

print(f"\nOn average, successful calls are {difference:.0f} seconds longer than unsuccessful calls.")
```
